```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('spi_matches.csv')
df
```

Biểu đồ cột :2016.10 đội ghi nhiều bàn thắng nhất trong một trận đấu là bao nhiêu bàn thắng?

```python
df_selected = df.head(10)
plt.figure(figsize=[10,3])
plt.bar(df_selected['season'],df_selected['league'],df_selected['score1'],df_selected['team1'],color='green')
plt.title('2016, 10 đội ghi nhiều bàn nhất trong một trận đấu',fontsize=10)
plt.xlabel('Tên đội bóng',fontsize=10)
plt.ylabel('Bàn thắng (Điểm)',fontsize=10)
plt.xticks(rotation=90)
plt.show()
```

```python


# Sắp xếp các đội theo chỉ số tấn công từ cao đến thấp
df_sorted = df.sort_values(by='attack_score', ascending=False)

# Lấy chỉ số tấn công của đội mạnh nhất (đội có chỉ số tấn công cao nhất)
top_team_score = df_sorted.iloc[0]['attack_score']

# Lấy chỉ số tấn công của 5 đội tiếp theo
next_five_teams_scores = df_sorted.iloc[1:6]['attack_score']

# Tính trung bình chỉ số tấn công của 5 đội tiếp theo
average_next_five = next_five_teams_scores.mean()

# Tính chênh lệch giữa đội mạnh nhất và trung bình của 5 đội tiếp theo
difference = top_team_score - average_next_five

print(f"Chênh lệch chỉ số tấn công giữa đội mạnh nhất và trung bình của 5 đội tiếp theo là: {difference:.2f}")
```

```python


# Lọc dữ liệu chỉ cho mùa giải 2019
df_2019 = df[df['season'] == 2019]

# Tìm đội có số bàn thắng nhiều nhất
top_team = df_2019.loc[df_2019['goals']]

# Lấy tên đội bóng và số bàn thắng
team_name = top_team['team']
goals_scored = top_team['goals']

print(f"Đội bóng ghi nhiều bàn thắng nhất trong mùa 2019 là {team_name} với {goals_scored} bàn.")
```
