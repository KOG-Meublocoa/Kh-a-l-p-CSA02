```python
# df_region = df.loc[["Region"]=="Đồng bằng sông Hồng"]
# df_region['Density']= df_region['Population']/df_region['Area']

df_region = df.groupby('Region').agg({
    'Area': 'sum',
    'Population': 'sum'      
}).reset_index()

df_region['Density'] = df_region['Population'] / df_region['Area']
df_region
```

```python
df_dense = df.groupby('Region')[['Population','Area']].sum()
df_dense['Density'] = df_dense['Population'] / df_dense['Area']
df_dense
```

```python

plt.figure(figsize=[10,6])
plt.bar(df_region['Region'],df_region['Density'],color='green')
plt.title('Mật độ dân số',fontsize=14)
plt.xlabel('Region',fontsize=12)
plt.ylabel('Density',fontsize=12)
plt.xticks(rotation=90)
plt.show()
```
